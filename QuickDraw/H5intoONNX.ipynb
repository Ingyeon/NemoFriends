```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import keras2onnx

model = load_model("./Model/QuickDraw_mobileNet.h5")

model.summary()
```

```python
onnx_model = keras2onnx.convert_keras(model,
                                        name="QuickDrawModel",
                                        target_opset=9,
                                        channel_first_inputs=None
                                      )

keras2onnx.save_model(onnx_model,"./Model/QuickDrawModel.onnx")
```
